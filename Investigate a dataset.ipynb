# Cleaning the Data

```python
#checking the number of duplicated rows

sum(df.duplicated())
```

```python
#dropping the duplicates

df.drop_duplicates(inplace = True)
```

```python
#checking again to confirm there are no duplicated rows

sum(df.duplicated())
```

* Duplicates removed

```python
#Making a copy of the original DataFrame to continue cleaning

df_cleaning = df.copy()
df_cleaning
```

```python
#dropping the unecessary rows

df_cleaning.drop(columns = ['imdb_id','budget', 'revenue','cast','homepage','director','tagline','keywords','overview','production_companies','budget_adj','revenue_adj'], inplace = True)
```

```python
#checking for null values

df_cleaning.isnull().sum()
```

There are **23** null values in the genres column

```python
#droppping the null values

df_cleaning.dropna(inplace = True)
```

```python
#checking again to confirm there are no null values

df_cleaning.isnull().sum()
```

* Null values removed

```python
#checking again the total number of non-null records in each row and the data types of the columns

df_cleaning.info()
```

# Q1. How does the number of movies per year changes throughout the time?

```python
#plotting the release year vs the no. of movies

ax_1 = df_cleaning.groupby('release_year').count().id.plot()

ax_1.set_xlabel('Release Year', labelpad = 20, size = 12)
ax_1.set_ylabel('No. of movies', labelpad = 20, size = 12);
ax_1.set_title('Number of movies per year though time', size = 16, pad = 25)
```

Wow, The number of the movies per year through the time is increasing exponentially!

# Q2. How do the averages of the votes for each of the movies and run time  spread?

```python
#plotting the histogram for the vote_average column

ax_2 = df_cleaning.vote_average.hist()

ax_2.set_xlabel('Average Votes', labelpad = 20, size = 12)
ax_2.set_ylabel('Frequency', labelpad = 20, size = 12)
ax_2.set_title('Average Votes Distribution', size = 16, pad = 25)
```

The average votes are nearly normally distributed around **5.5**

```python
#plotting the histogram for the run time column

ax_2 = df_cleaning.runtime.hist()

ax_2.set_xlabel('Run Time', labelpad = 20, size = 12)
ax_2.set_ylabel('Frequency', labelpad = 20, size = 12)
ax_2.set_title('Run Time Distribution', size = 16, pad = 25)
```

The Run time is nearly concentrated at **150**:**190** min.

# Q4. Ok... Let's see what is the most popular category? 

```python
#plotting the popularity of each category

ax_4 = df_c0.groupby('genres').mean().popularity.sort_values(ascending = False).plot(kind = 'bar')

ax_4.set_xlabel('genres', size = 12, labelpad = 20)

ax_4.set_ylabel('popularity', size = 12, labelpad = 20)

ax_4.set_title('Popularity Vs Genre', size = 16, pad = 25)
```

It's the **ADVENTURE** movies that has the highest average popularity.

# Q5. Does this mean "Adventure Movies" have the highest average votes?
